```julia
power(x,y,s) = let rackid = x + 10
    ((rackid * y + s) * rackid ÷ 100) % 10 - 5
end

@show power(3,5,8)
@show power(122, 79, 57)
@show power(217, 196, 39)
```

```julia
const serial = 9005
const S = 300
P = reshape([power(x,y,serial) for y in 1:S, x in 1:S], (S,S))
```

```julia
# checker
@show P[3,1]  # x,y are reversed so x=1, y=3 
@show power(3,1,serial)
```

```julia
Q2 = [sum(P[j:j+2,i:i+2]) for j in 1:297, i in 1:297]
```

```julia
findmax(Q2)
```

```julia
Q(z,S) = findmax([sum(P[j:j+z-1,i:i+z-1]) for i in 1:S-z, j in 1:S-z])
Q(3,S)
```

```julia
using Dates: now
for i in 1:S-1
    println(string(now())[1:19], "   ", i, " => ", Q(i, S))
end
```

```julia
using Plots

problem_size(cell_size, grid_size) = cell_size * cell_size * (grid_size - cell_size + 1) * (grid_size - cell_size + 1)
plot(1:S, problem_size.(1:S, S))
```
